```python
# use a seaborn map to find all the NaN values (1pt)

cols = df.columns
sns.heatmap(df[cols].isnull())
```

```python
# replace those NaN values (1pt)
df_copy = df.copy()
df_copy["last_review"] = df_copy["last_review"].fillna("Not Specified")
df_copy["reviews_per_month"] = df_copy["reviews_per_month"].fillna("Not Specified")
df_copy
```

```python
# isolate just the year from the last_review column (1pt)
df_copy.last_review = df_copy.last_review.str[0:4]
df_copy.last_review
for x in df_copy.last_review:
    if x == "Not ":
        continue
    else:
        print(x)
```

```python
# and do one other thing with the dataset of your choosing (1pt)
neighborhoods = df_copy["neighbourhood"]

list_of_neighborhoods = []
for x in neighborhoods:
    if x not in list_of_neighborhoods:
        list_of_neighborhoods.append(x)

print(list_of_neighborhoods)

# prints a list of all the neighborhoods with no repeat values
```
